```python

import os
from google import genai
from google.genai.types import GenerateContentConfig
from google.colab import userdata   # ← Added for Colab Secrets
```

```python
# ==================== API Key Setup for Colab ====================
# Go to: Runtime → Secrets (🔑) and add a secret named GEMINI_API_KEY
GEMINI_API_KEY = userdata.get('GEMINI_API_KEY')

client = genai.Client(api_key=GEMINI_API_KEY)

MODEL = "gemini-2.5-flash"
```

```python

  # ==================== Core Function ====================
def run_agent(system_prompt: str, user_prompt: str, temperature=0.7) -> str:
    """Simple reusable agent call"""
    response = client.models.generate_content(
        model=MODEL,
        contents=[
            {"role": "user", "parts": [{"text": system_prompt}]},
            {"role": "user", "parts": [{"text": user_prompt}]}
        ],
        config=GenerateContentConfig(
            temperature=temperature,
            max_output_tokens=2048,
        )
    )
    return response.text.strip()
```

```python
# ==================== Agents ====================

def researcher(topic: str) -> str:
    print(f"\n🔍 RESEARCHER — '{topic}'\n" + "─" * 60)
    system = "You are a research agent. Return exactly 5 concise bullet-point facts."
    user = f"Research this topic: {topic}"
    result = run_agent(system, user)
    print(result)
    return result


def critic(research: str) -> str:
    print(f"\n🧐 CRITIC\n" + "─" * 60)
    system = "You are a critical reviewer. Identify 2-3 gaps, errors, or improvements."
    user = f"Review these research notes:\n{research}"
    result = run_agent(system, user)
    print(result)
    return result


def writer(research: str, critique: str) -> str:
    print(f"\n✍️ WRITER\n" + "─" * 60)
    system = "You are a professional writer. Write a clean, polished 3-sentence summary using the research and addressing the critique."
    user = f"Research:\n{research}\n\nCritique:\n{critique}"
    result = run_agent(system, user)
    print(result)
    return result
```

```python
# ==================== Pipeline ====================
def run_pipeline(topic: str):
    print(f"🚀 Starting Agent Pipeline for: {topic}\n")

    notes = researcher(topic)
    feedback = critic(notes)
    summary = writer(notes, feedback)

    print("\n" + "="*70)
    print("✅ Pipeline Complete!")

    return {
        "research": notes,
        "critique": feedback,
        "summary": summary
    }


```

```python
# ==================== Run ====================
if __name__ == "__main__":
    result = run_pipeline("The future of quantum computing")
    # Optional: display final summary nicely
    print("\n📝 Final Summary:")
    print(result["summary"])
```
